```python
df['size'] = df['size'].fillna('2 BHK')
df['bath'] = df['bath'].fillna(df['bath'].median())
```

```python
df['bhk']=df['size'].str.split().str.get(0).astype(int)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
```

```python
df.head()
```

```python
df['size'].unique()
```

```python
df['price_per_sqft'] = df['price']*100000/df['total_sqft']
df['price_per_sqft']
```

```python
df['location'].value_counts()
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
location_count = df['location'].value_counts()
location_count
```

```python
location_count_less_10 = location_count[location_count<=10]
location_count_less_10
```

```python
df['location'] = df['location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

```python
(df['total_sqft']/df['bhk']).describe()
```

```python
df = df[((df['total_sqft']/df['bhk'])>=300)]
```

```python
df.shape
```

```python
def remove_outliers_sqft(df2):
  df2_output = pd.DataFrame()
  for key, subdf in df2.groupby('location'):
    m = np.mean(subdf.price_per_sqft)

    st = np.std(subdf.price_per_sqft)

    gen_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
    df2_output = pd.concat([df2_output,gen_df],ignore_index=True)
  return df2_output
  df = remove_outliers_sqft(df2)
```

```python
df.describe()
```

```python
def bhk_outlier_remover(df3):
  exclude_indices = np.array([])
  for location, location_df in df3.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std': np.std(bhk_df.price_per_sqft),
          'count': bhk_df.shape[0]
      }

    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df3.drop(exclude_indices,axis='index')
```

```python
df = bhk_outlier_remover(df)
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
trs1 = make_column_transformer((OneHotEncoder(sparse=False), ['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
lr = Ridge(alpha=True)
```

```python
pipe = make_pipeline(trs1,scaler,lr)
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
pipe2 = make_pipeline(trs1,DecisionTreeRegressor())
pipe2.fit(x_train,y_train)
```

```python
pipe2.predict(x_test)
```

```python
r2_score(y_test,pipe2.predict(x_test))
```

```python
lasso = Lasso()
pipe3 = make_pipeline(trs1,scaler,lasso)
pipe3.fit(x_train,y_train)
```

```python
r2_score(y_test,pipe3.predict(x_test))
```
